```python
import os
import gymnasium as gym
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv, VecMonitor, VecVideoRecorder
from stable_baselines3.common.monitor import Monitor
from stable_baselines3.common.callbacks import EvalCallback

def make_env():
    env = gym.make("BipedalWalker-v3")
    env = Monitor(env)
    return env

env = DummyVecEnv([make_env])
env = VecMonitor(env)
```

```python
RUN_DIR = "run/bipedal_baseline"
# VIDEO_DIR = os.path.join(RUN_DIR, "video")
# os.makedirs(VIDEO_DIR, exist_ok=True)
BEST_DIR = os.path.join(RUN_DIR, "best")
os.makedirs(BEST_DIR, exist_ok=True)
LOG_DIR = os.path.join(RUN_DIR, "log")
os.makedirs(LOG_DIR, exist_ok=True)
```

```python
def make_train_env():
  env = gym.make("BipedalWalker-v3", render_mode = "rgb_array")
  env = RewardWrapper(
    env,
    action_penalty_coefficient = 0.001,
    alive_bonus = 0.0
  )
  env = Monitor(env)
  return env
def make_eval_env():
  env = gym.make("BipedalWalker-v3", render_mode = "rgb_array")
  env = RewardWrapper(
      env,
      action_penalty_coefficient = 0.001,
      alive_bonus = 0.0
  )
  env = Monitor(env)

  return env
```

```python
train_env = DummyVecEnv([make_train_env])
train_env = VecMonitor(train_env)

eval_env = DummyVecEnv([make_eval_env])
eval_env = VecMonitor(eval_env)
```

```python
eval_env = VecVideoRecorder(
    eval_env,
    VIDEO_DIR,
    record_video_trigger=lambda step: True,
    video_length=1600,
    name_prefix="eval",
)
```

```python
eval_callback = EvalCallback(
    eval_env,
    best_model_save_path = BEST_DIR,
    log_path = LOG_DIR,
    eval_freq = 50_000,
    n_eval_episodes = 5,
    deterministic=True,
    render=False
)
```

```python
model = PPO(
    "MlpPolicy",
    train_env,
    learning_rate = 3e-4,
    n_steps = 2048,
    batch_size = 64,
    n_epochs = 10,
    gamma = 0.99,
    gae_lambda = 0.95,
    clip_range=0.2,
    ent_coef = 0.0,
    verbose = 1,
    tensorboard_log = os.path.join(RUN_DIR, "tensorboard"),
)
```

```python
model.learn(
    total_timesteps = 300_000,
    callback = eval_callback,
    progress_bar = False
)
```

```python
model.save(os.path.join(RUN_DIR, "model"))
train_env.close()
eval_env.close()
print("Training complete.")
print("Videos saved to:", VIDEO_DIR)
print("Best model saved to:", BEST_DIR)
```

```python
import gymnasium as gym
import numpy as np

from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv, VecNormalize


class RewardWrapper(gym.Wrapper):
  def __init__(self, env, action_penalty_coefficient = 0.001, alive_bonus = 0.0):
    super().__init__(env)
    self.action_penalty_coefficient = action_penalty_coefficient
    self.alive_bonus = alive_bonus

  def step(self, action):
    obs, reward, terminated, truncated, info = self.env.step(action)
    shaped_reward = reward
    action_penalty = self.action_penalty_coefficient * float(np.mean(np.square(action)))
    shaped_reward -= action_penalty
    shaped_reward += self.alive_bonus
    info = dict(info)
    info["reward_original"] = float(reward)
    info["reward_shaped"] = float(shaped_reward)
    return obs, shaped_reward, terminated, truncated, info
```
